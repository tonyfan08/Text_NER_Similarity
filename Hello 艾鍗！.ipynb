```python
# 載入同義字字典
combine_dict = {}
# Replace_Words.txt是同義詞表，每行是一系列同義詞，用空格分割
for line in open("/content/drive/MyDrive/test/Keywords/Replace_Words.txt", "r", encoding='utf-8'):
    seperate_word = line.strip().split(" ")
    num = len(seperate_word)
    for i in range(1, num):
        combine_dict[seperate_word[i]] = seperate_word[0]
```

```python
train['tokens']=""
for i in range(train.shape[0]):
  tokens = [token for token in jieba.lcut(train['text'][i])if token!=' ']

  # 將分詞中的同義詞，替換成字典中的第一個字
  for idx, word in enumerate(tokens):
      if word in combine_dict:
        tokens[idx] = combine_dict[word]
  train['tokens'][i]=tokens
train.head()
```

##cosine similarity

####各關鍵字分各關鍵字分開

```python
k=0
for i in range(train.shape[0]):
  for j in range(i+1,train.shape[0]):
    k+=1
print(k)
```

```python
ddf = pd.DataFrame(columns=['Test','Reference','comb','sim_value_chem','sim_value_crop','sim_value_pest','sim_value_other','len1','len2','len3','len4','Label'],index=list(range(155961)))
```

```python
txtnumlist=[i for i in train['file_name']]#df['file_name'].values.tolist()
txtnumlist=[int(i.replace('.txt','')) for i in txtnumlist]
```

```python
#每兩句的cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
k=0
for i in range(train.shape[0]):
  for j in range(i+1,train.shape[0]):
    a = dfr_chem.values[i]
    b = dfr_chem.values[j]
    c = dfr_crop.values[i]
    d = dfr_crop.values[j]
    e = dfr_pest.values[i]
    f = dfr_pest.values[j]
    g = dfr_other.values[i]
    h = dfr_other.values[j]

    #Dot and norm
    Dot1 = np.dot(a,b)
    norm_a = np.dot(a,a)**0.5
    norm_b = np.dot(b,b)**0.5
    Dot2 = np.dot(c,d)
    norm_c = np.dot(c,c)**0.5
    norm_d = np.dot(d,d)**0.5
    Dot3 = np.dot(e,f)
    norm_e = np.dot(e,e)**0.5
    norm_f = np.dot(f,f)**0.5
    Dot4 = np.dot(g,h)
    norm_g = np.dot(g,g)**0.5
    norm_h = np.dot(h,h)**0.5

    #Cosine similarity
    cos_sim1 = Dot1 / (norm_a*norm_b)
    ddf['sim_value_chem'][k]=cos_sim1
    cos_sim2 = Dot2 / (norm_c*norm_d)
    ddf['sim_value_crop'][k]=cos_sim2
    cos_sim3 = Dot3 / (norm_e*norm_f)
    ddf['sim_value_pest'][k]=cos_sim3
    cos_sim4 = Dot4 / (norm_g*norm_h)
    ddf['sim_value_other'][k]=cos_sim4

    #上Test.Reference
    ddf['Test'][k]=txtnumlist[i]
    ddf['Reference'][k]=txtnumlist[j]
    k+=1
```

```python
#交換順序
back=ddf[['Reference','Test','comb','sim_value_chem','sim_value_crop','sim_value_pest','sim_value_other','len1','len2','len3','len4','Label']]
back.head(10)
```

```python
#重新命名
back.rename(columns={'Reference':'Test','Test':'Reference'},inplace=True)
back
```

```python
#連接
ddf_sim_value = pd.concat([ddf,back])
ddf_sim_value.reset_index(drop=True,inplace=True)
ddf_sim_value
```

```python
ddf_sim_value['comb']=ddf_sim_value['Test'].astype(str)+'_'+ddf_sim_value['Reference'].astype(str)
ddf_sim_value
```

##cosine similarity

####各關鍵字分各關鍵字分開

```python
k=0
for i in range(Private.shape[0]):
  for j in range(i+1,Private.shape[0]):
    k+=1
print(k)
```

```python
cos_sim_df = pd.DataFrame(columns=['Test','Reference','comb','sim_value_chem','sim_value_crop','sim_value_pest','sim_value_other','len1','len2','len3','len4'],index=list(range(87990)))
```

```python
txtnumlist=[i for i in Private['file_name']]#df['file_name'].values.tolist()
txtnumlist=[i.replace('.txt','') for i in txtnumlist]
```

```python
#每兩句的cosine_similarity
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
k=0
for i in range(Private.shape[0]):
  for j in range(i+1,Private.shape[0]):
    a = Private_chem.values[i]
    b = Private_chem.values[j]
    c = Private_crop.values[i]
    d = Private_crop.values[j]
    e = Private_pest.values[i]
    f = Private_pest.values[j]
    g = Private_other.values[i]
    h = Private_other.values[j]

    #Dot and norm
    Dot1 = np.dot(a,b)
    norm_a = np.dot(a,a)**0.5
    norm_b = np.dot(b,b)**0.5
    Dot2 = np.dot(c,d)
    norm_c = np.dot(c,c)**0.5
    norm_d = np.dot(d,d)**0.5
    Dot3 = np.dot(e,f)
    norm_e = np.dot(e,e)**0.5
    norm_f = np.dot(f,f)**0.5
    Dot4 = np.dot(g,h)
    norm_g = np.dot(g,g)**0.5
    norm_h = np.dot(h,h)**0.5

    #Cosine similarity
    cos_sim1 = Dot1 / (norm_a*norm_b)
    cos_sim_df['sim_value_chem'][k]=cos_sim1
    cos_sim2 = Dot2 / (norm_c*norm_d)
    cos_sim_df['sim_value_crop'][k]=cos_sim2
    cos_sim3 = Dot3 / (norm_e*norm_f)
    cos_sim_df['sim_value_pest'][k]=cos_sim3
    cos_sim4 = Dot4 / (norm_g*norm_h)
    cos_sim_df['sim_value_other'][k]=cos_sim4

    #上Test.Reference
    cos_sim_df['Test'][k]=txtnumlist[i]
    cos_sim_df['Reference'][k]=txtnumlist[j]
    k+=1
```

```python
#交換順序
back=cos_sim_df[['Reference','Test','comb','sim_value_chem','sim_value_crop','sim_value_pest','sim_value_other','len1','len2','len3','len4']]
back.head(10)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

#Training with function, set a K value loop
Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))

for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

#show the best model
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1) 
```

```python
import joblib
neigh = joblib.load('neigh1_1213.model')
```

```python
# Classifying new data
#test_data = pd.read_pickle("testpublicdata_1208")  # read test data
test_input = np.array(concatdf.iloc[:,3:6])
```

```python
yhat=neigh.predict(test_input)
predict_class = le.inverse_transform(yhat)
df_predict= pd.DataFrame(predict_class , columns=['Label'])
df_predict['Test']=concatdf['Test']
df_predict['Reference']=concatdf['Reference']
df_predict.head()
```

```python
#挑出yes的
a = df_predict[df_predict['Label']=='Yes']
a["comb"]=a["Test"].astype(str)+"_"+a["Reference"].astype(str)
a.reset_index(drop=True, inplace=True)
a
```

###篩選cos_sim>0.7

```python
concatdf["mark"]=""
for i in a["comb"]:
  for j in concatdf['comb']:
    if j==i:
      Num = concatdf[concatdf['comb']==j].index.item()
      concatdf["mark"][Num]=1
concatdf
```

```python
prepare = concatdf[concatdf['mark']==1]
b = prepare[prepare['sim_value']>0.7]
c = b[["Test","Reference"]]
c
```

```python
c.to_csv('knn_sim>0.7.csv',index=False)
```
